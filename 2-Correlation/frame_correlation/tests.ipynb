```python
sample_size = 5*10**5
print(sample_size)
plt.plot(data[0:sample_size])
```

```python
def mean(X):
    return np.sum(X, axis=0)/len(X)

def std_dev(X, X_bar):
    return np.sqrt(np.sum((X-X_bar)**2, axis=0))

def cov(X, X_bar, Y, Y_bar):
    return np.sum((X-X_bar)*(Y-Y_bar), axis=0)

def pearson(X,Y):
    X_mean = mean(X)
    Y_mean = mean(Y)    
    return cov(X, X_mean, Y, Y_mean)/(std_dev(X, X_mean)*std_dev(Y, Y_mean))
```

```python
samp1 = data[0:sample_size]
samp2 = data[sample_size:sample_size*2]
```

```python
r = pearson(samp1, samp2)
print(r)
```
